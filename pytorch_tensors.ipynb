### Transpose

This switches the dimension of a tensor

Suppose, original tensore has `(540, 768, 3)` shape with dimension of 3. For example this could be `(image_width, image_ht, colour_channels)`

* Along dim=0, shape is `540`
* Along dim=1, shape is `768`
* Along dim=2, shape is `3`

If we want to transform our data so, that resulting shape should be `(color_channels, image_ht, image_width)`. Then `torch.transpose` can be used to switch shape by interchanging 2 dimensions

```python
random_image_size_tensor = torch.rand(size=(540,768,3))
(torch.Size([540, 768, 3]), 3)

random_image_size_tensor_transposed = random_image_size_tensor.transpose(dim0=0, dim1=2)
torch.Size([3, 768, 540])
```



````python
random_image_size_tensor = torch.rand(size=(540,768,3))
random_image_size_tensor.shape, random_image_size_tensor.ndim
````
